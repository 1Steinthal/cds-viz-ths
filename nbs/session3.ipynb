````python
#Original filepath
filepath = os.path.join("..", "data", "img")

#Function that creates colorhistogram for each file in given path and saves it
def create_image_hist(path):
    colors = ('b','g','r') 
    imgs = sorted(os.listdir(path)) #list of images

    for img in imgs:
        imgpath = os.path.join(path, img)
        image = cv2.imread(imgpath)
        for i,color in enumerate(colors): 
            hist = cv2.calcHist([image],[i],None,[256],[0,256]) 
            plt.plot(hist,color = color) 
        plt.title("Histogram of " + img) 
        plt.xlabel('Pixel value')
        plt.ylabel('# of pixels')
        plt.legend('BGR')

        print(img)
        #Saving image
        saveas = os.path.join("..", "output", "session3", "BGRhistogramOf_" + img)
        plt.savefig(saveas)
        print("Image saved")
        plt.clf()
        print("plot cleared")



create_image_hist(filepath)
        

````

# Comparing histograms

````python
filepath_1 = os.path.join("..", "data", "img", "trex.png")
filepath_2 = os.path.join("..", "data", "img", "t-rex2.png")

img_1 = cv2.imread(filepath_1)
img_2 = cv2.imread(filepath_2)
````

````python
hist_1 = cv2.calcHist([img_1], [0,1,2], None, [255,255,255], [0,256,0,256,0,256])
hist_2 = cv2.calcHist([img_2], [0,1,2], None, [255,255,255], [0,256,0,256,0,256])
````

````python
normalized_hist_1 = cv2.normalize(hist_1, hist_1, 0, 1.0, cv2.NORM_MINMAX)
normalized_hist_2 = cv2.normalize(hist_2, hist_2, 0, 1.0, cv2.NORM_MINMAX)
````

````python
cv2.compareHist(normalized_hist_1, normalized_hist_2, cv2.HISTCMP_BHATTACHARYYA)
````
